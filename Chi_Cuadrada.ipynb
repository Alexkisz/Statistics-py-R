```r
# Crear una matriz de datos de ejemplo
datos <- matrix(c(30, 20, 15, 25, 10, 40, 20, 30, 5), nrow = 3, byrow = TRUE)
colnames(datos) <- c("Profesional", "Técnico", "No calificado")
rownames(datos) <- c("Educación Primaria", "Educación Secundaria", "Educación Superior")

# Mostrar la tabla de contingencia
print(datos)

# Realizar la prueba chi-cuadrado
resultado_chi_cuadrado <- chisq.test(datos)

# Mostrar el resultado de la prueba
print(resultado_chi_cuadrado) ## 0.00000002501

## Por lo tanto, hay evidencia estadística significativa para concluir que hay una asociación significativa entre el nivel educativo y el tipo de ocupación en la muestra.

# Graficar la tabla de contingencia
barplot(t(datos), beside = TRUE, legend.text = TRUE, col = rainbow(3),
        main = "Frecuencia de ocupación por nivel educativo",
        xlab = "Nivel educativo", ylab = "Frecuencia", ylim = c(0, 80))

# Agregar la leyenda
legend("topright", legend = colnames(datos), fill = rainbow(3))


```
