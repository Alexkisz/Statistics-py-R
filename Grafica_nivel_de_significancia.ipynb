```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Definir los grados de libertad
df1 = 10  # Grados de libertad del numerador
df2 = 15  # Grados de libertad del denominador

# Generar valores para la distribución F
x = np.linspace(0.01, 5, 1000)
y = f.pdf(x, df1, df2)

# Calcular el valor crítico para un nivel de significancia dado
alpha = 0.05
critical_value = f.ppf(1 - alpha, df1, df2)

# Graficar la distribución F
plt.plot(x, y, 'b-', label='Distribución F')
plt.fill_between(x[x > critical_value], 0, y[x > critical_value], color='red', alpha=0.5, label='Región Crítica')
plt.xlabel('Valor del Estadístico F')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución F con Región Crítica')
plt.legend()
plt.grid(True)
plt.show()
```
